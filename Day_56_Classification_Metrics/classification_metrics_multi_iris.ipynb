```python
df = pd.read_csv('/kaggle/input/iris/Iris.csv')
```

```python
df.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Species'] = encoder.fit_transform(df['Species'])
```

```python
df.sample(5)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1], test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

lor=LogisticRegression()
dtc=DecisionTreeClassifier()
```

```python
lor.fit(X_train,y_train)
dtc.fit(X_train,y_train)

y_pred_lor = lor.predict(X_test)
y_pred_dtc = dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Logistic Regression Prediction'] = y_pred_lor
result['Decision Tree Prediction'] = y_pred_dtc

result.sample(10)
```

```python
print("Logistic Regression Confusion Matrix")
df = pd.DataFrame(confusion_matrix(y_test,y_pred_lor), columns=list(range(0,3)))
print(df)
print("*"*50)
print("Accuracy: ", accuracy_score(y_test,y_pred_lor))
print("Precision: ", precision_score(y_test, y_pred_lor,average='weighted'))
print("Recall: ", recall_score(y_test,y_pred_lor,average='weighted'))
print("F1 Score: ", f1_score(y_test, y_pred_lor,average='macro'))
      
```

```python
print("Decision Tree Confusion Matrix")
df = pd.DataFrame(confusion_matrix(y_test,y_pred_lor), columns=list(range(0,3)))
print(df)
print("*"*50)
print("Accuracy: ", accuracy_score(y_test,y_pred_lor))
print("Precision: ", precision_score(y_test, y_pred_dtc,average='macro'))
print("Recall: ", recall_score(y_test,y_pred_dtc,average='macro'))
print("F1 Score: ", f1_score(y_test, y_pred_dtc,average='macro'))
```

```python
df = pd.read_csv('/kaggle/input/mnist-digit-recognizer/train.csv')
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

lor = LogisticRegression()
dtc = DecisionTreeClassifier()

lor.fit(X_train, y_train)
dtc.fit(X_train,y_train)
```

```python
y_pred_lor = lor.predict(X_test)
y_pred_dtc = dtc.predict(X_test)
```

###### from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score


```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, classification_report
```

```python
print("Logistic Regression Confusion Matrix and Metrics")

df = pd.DataFrame(confusion_matrix(y_test,y_pred_lor), columns=list(range(0,10)))
print(df)
print("-"*50)

print("Accuracy: ", accuracy_score(y_test, y_pred_lor))
print("Precision: ", precision_score(y_test, y_pred_lor, average='weighted'))
print("Recall: ", recall_score(y_test, y_pred_lor, average='weighted'))
print("F1 Score: ", f1_score(y_test, y_pred_lor, average='weighted'))

print("-"*50)
print("Classification Report")
print(classification_report(y_test, y_pred_lor))
```

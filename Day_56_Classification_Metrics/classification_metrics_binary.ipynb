```python
df = pd.read_csv('/kaggle/input/heart-diseaseuci/heart.csv')
print(df.shape)
df.sample(10)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
lor = LogisticRegression()
dtc = DecisionTreeClassifier()
```

```python
lor.fit(X_train,y_train)
dtc.fit(X_train,y_train)
```

```python
y_pred_lor = lor.predict(X_test)
y_pred_dtc = dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print("Accuracy of Logistic Regression: ", accuracy_score(y_test,y_pred_lor))
print("Accuracy of Logistic Regression: ", accuracy_score(y_test,y_pred_dtc))
```

```python
print("Logistic Regression Confusion Matrix")
pd.DataFrame(confusion_matrix(y_test,y_pred_lor), columns=list(range(0,2)))
```

```python
print("Decision Tree Confusion Matrix")
pd.DataFrame(confusion_matrix(y_test,y_pred_dtc), columns=list(range(0,2)))
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Logistic Regression Prediction'] = y_pred_lor
result['Decision Tree Prediction'] = y_pred_dtc
result.sample(10)
```

```python
from sklearn.metrics import recall_score, precision_score, f1_score
```

```python
print("For Logistic Regression Model")
print("-"*50)

cmf = pd.DataFrame(confusion_matrix(y_test,y_pred_lor), columns=list(range(0,2)))
print(cmf)

print("-"*50)
print("Precision : ", precision_score(y_test, y_pred_lor))
print("Recall : ", recall_score(y_test, y_pred_lor))
print("F1 Score : ", f1_score(y_test, y_pred_lor))
```

```python
print("Decision Tree Model")

print("-"*50)
cmf = pd.DataFrame(confusion_matrix(y_test,y_pred_dtc),columns=list(range(0,2)))
print(cmf)
print("-"*50)

print("Precision : ", precision_score(y_test,y_pred_dtc))
print("Recall : ", recall_score(y_test, y_pred_dtc))
print("F1 Score : ", f1_score(y_test, y_pred_dtc))
```
